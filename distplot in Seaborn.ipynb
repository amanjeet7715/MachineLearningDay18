```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

**distplot is plotted**

```python
sns.distplot(iris['sepal_length'])
```

**bins is used to set the number of bins we want in our plot and it actually depends on our dataset**

```python
sns.distplot(iris['sepal_width'], bins=10)
```

**hist is used for Whether to plot a (normed) histogram**

```python
sns.distplot(iris['sepal_width'], hist=False)
```

**KDE stands for Kernel Density Estimation and that is another kind of the plot in seaborn**

```python
sns.distplot(iris['sepal_width'], kde=False)
```

**rug is used for whether to draw a rugplot on the support axis**

```python
sns.distplot(iris['sepal_width'], rug=True)
```

**An object with fit method, returning a tuple that can be passed to a pdf method a positional arguments following a grid of values to evaluate the pdf on**

```python
sns.distplot(iris['sepal_width'], fit=norm, kde=False)
```

**color is used to specify the color of the plot**

```python
sns.distplot(iris['sepal_width'], color='red')
```

**vertical
If True, observed values are on y-axis**

```python
sns.distplot(iris['sepal_width'], vertical=True)
```

**norm_hist
If True, the histogram height shows a density rather than a count. This is implied if a KDE or fitted density is plotted**

```python
sns.distplot(iris['sepal_width'], norm_hist=True)
```

**axlabel**
Name for the support axis label. If None, will try to get it from a.name if False, do not set a label

```python
sns.distplot(iris['sepal_width'], axlabel='sepal_wid')
```

**label sets Legend label for the relevant component of the plot**

```python
sns.distplot(iris['sepal_width'], label='sepal')
plt.legend()
```
